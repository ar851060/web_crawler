```python
import requests
import pandas as pd
switchs = []
for page in range(1,6):
    res = requests.get(f'https://search.taobao.tw/catalog/?_keyori=ss&ajax=true&from=input&page={page}&q=switch')
    jd = res.json()
    df = pd.DataFrame(jd['mods']['listItems'])
    switchs.append(df)
```

```python
df_all = pd.concat(switchs)
len(df_all)
```
